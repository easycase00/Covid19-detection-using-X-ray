```python
import pandas as pd
import os
import shutil
```

## Creating dataset for +ve samples

```python
File_path= 'C:/Users/Admin/PycharmProjects/Covid-19-detection-using-X-rays/covid-chestxray-dataset-master/metadata.csv'
Image_path= 'C:/Users/Admin/PycharmProjects/Covid-19-detection-using-X-rays/covid-chestxray-dataset-master/images/'
```

```python
df= pd.read_csv(File_path)
df.shape
```

```python
df.head()
```

```python
Target_dir= 'Dataset/Covid'

if not os.path.exists(Target_dir):
    os.mkdir(Target_dir)
    print("folder created")
else:
    print('already created')
```

```python
counter = 0

for (i,row) in df.iterrows() :
    if row["finding"]=="Pneumonia/Viral/COVID-19" and row["view"]=="PA":
        filename= row["filename"]
        image_path= os.path.join(Image_path, filename)
        image_copy_path= os.path.join(Target_dir, filename)
        shutil.copy2(image_path, image_copy_path)
        counter += 1
        print("moving image", counter)
```

## Making dataset for normal/-ve images

```python
import random
Kaggle_file_path= "chest_xray/chest_xray/train/NORMAL"
Target_normal_dir= "Dataset/Normal"
```

```python
image_names= os.listdir(Kaggle_file_path)

random.shuffle(image_names)
print(image_names)
```

```python

for i in range(196):
    image_name= image_names[i]
    print(image_name)
    
    image_path= os.path.join(Kaggle_file_path, image_name)
    
    target_path= os.path.join(Target_normal_dir, image_name)
    
    shutil.copy2(image_path, target_path)
    
    print("copying image", i)
```

## Covid Train and Test Data
